```python
import pandas as pd
```

```python
csv1 = r"E:\Projetos\Github_ViniciusJ\Pesquisa_IC_IA\gerdau_normalizado_v1.csv"
v1 = pd.read_csv(csv1)

csv2 = r"E:\Projetos\Github_ViniciusJ\Pesquisa_IC_IA\gerdau_normalizado_v2.csv"
v2 = pd.read_csv(csv2)
```

```python
# Corrigir problema: separar DividendosProducao
if "DividendosProducao" in v2.columns:
    v2["Dividendos"] = v2["DividendosProducao"] 
    v2["Producao"] = "Não informado" 
    v2 = v2.drop(columns=["DividendosProducao"])
```

```python
all_cols = sorted(set(v1.columns) | set(v2.columns))
v1 = v1.reindex(columns=all_cols, fill_value="Não informado")
v2 = v2.reindex(columns=all_cols, fill_value="Não informado")
```

```python
df_unificado = pd.concat([v1, v2], ignore_index=True)
```

```python
output_file = "gerdau_normalizado_unificado.csv"
df_unificado.to_csv(output_file, index=False, encoding="utf-8")
```

```python
print("Dataset unificado criado:", df_unificado.shape)
print("Colunas finais:", df_unificado.columns.tolist())
```
